```python
# 1402 星空之夜
# 用一个哈希函数来表示一个块的值
W = int(input())
H = int(input())
g = []
locate = []
size = 0
sumStock = 0
hashs = []

for _ in range(H):
    g.append(list(input()))

def dfs(x, y):
    global size
    size += 1
    g[x][y] = '0'
    locate.append([x, y])
    for i in range(x - 1, x + 2):
        for j in range(y - 1, y + 2):
            if i >=0 and i < H and j >= 0 and j < W and g[i][j] == '1':
                dfs(i, j)

def get_hash():
    code = 0
    for i in range(size):
        for j in range(i + 1, size):
            x1, y1 = locate[i]
            x2, y2 = locate[j]
            code += ((x1 - x2) ** 2 + (y1 - y2) ** 2) ** 0.5
    return code

def dealg(m):
    for i in range(len(locate)):
        x, y = locate[i]
        g[x][y] = chr(ord('a') + m)

def get_label(bufhash):
    global sumStock
    label = -1
    for k in range(sumStock):
        if abs(bufhash - hashs[k]) < 1e-6:
            label = k
            break
    if label == -1:
        label = sumStock
        sumStock += 1
        hashs.append(bufhash)
    return label

for i in range(H):
    for j in range(W):
        if g[i][j] == '1':
            locate = []
            size = 0
            dfs(i, j)
            bufhash = get_hash()
            label = get_label(bufhash)
            dealg(label)
for i in range(H):
    print(''.join(g[i]))
```
