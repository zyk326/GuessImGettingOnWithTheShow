```python
# 最大公约数
# 4199 公约数
a, b = map(int, input().split())

q = int(input())

nu = []
ans = []
lenth = 0

def gcd(a, b):
    return gcd(b, a % b) if b else a

def init_divisors(a, b):
    global lenth
    global nu
    g = gcd(a, b)
    i = 1
    while(i <= g // i):
        if g % i == 0:
            nu.append(i)
            lenth += 1
            if i != g // i:
                nu.append(g // i)
                lenth += 1
        i += 1
    nu = sorted(nu)

init_divisors(a, b)

while(q):
    flag = 0
    l, r = map(int, input().split())
    L, R = 0, lenth - 1
    # for i in range(lenth - 1, -1, -1):
    #     if nu[i] >= l and nu[i] <= r:
    #         ans.append(nu[i])
    #         flag = 1
    #         break
    # if flag ^ 1:
    #     ans.append(-1)
    while(L < R):
        mid = (L + R + 1) >> 1
        if nu[mid] <= r:
            L = mid
        else:
            R = mid - 1
    if nu[L] >= l:
        ans.append(nu[L])
    else:
        ans.append(-1)
    q -= 1

for i in ans:
    print(i)
```
