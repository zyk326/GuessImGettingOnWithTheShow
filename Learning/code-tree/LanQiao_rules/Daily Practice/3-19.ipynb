```python
# 687 扫雷
# Flood Fill 算法 低洼地注水算法，遇到第一个高低时 变值并返回返回
ans = []
T = int(input())
N = 310
str = [[' ' for i in range(N)] for j in range(N)]
g = [[0 for i in range(N)] for j in range(N)]
n = 0

def dfs(i, j):
    t = g[i][j]
    g[i][j] = -1
    if t != 0:
        return 
    for a in range(i - 1, i + 2):
        for b in range(j - 1, j + 2):
            if a >= 0 and a < n and b >= 0 and b < n and g[a][b] != -1:
                dfs(a, b)

while(T):
    str = [[' ' for i in range(N)] for j in range(N)]
    g = [[0 for i in range(N)] for j in range(N)]
    n = int(input())
    for i in range(n):
        line = input()
        for j in range(n):
            str[i][j] = line[j]

    for i in range(n):
        for j in range(n):
            if str[i][j] == '*':
                g[i][j] = -1
            else:
                for a in range(i - 1, i + 2):
                    for b in range(j - 1, j + 2):
                        if str[a][b] == '*':
                            g[i][j] += 1
    res = 0
    for i in range(n):
        for j in range(n):
            if g[i][j] == 0:
                res += 1
                dfs(i, j)
                
    for i in range(n):
        for j in range(n):
            if g[i][j] != -1:
                res += 1
    ans.append(res)
    T -= 1
for i in range(len(ans)):
    print("Case #%d: %d"%(i + 1, ans[i]))
```

```python
# 643 动态网格
T = int(input())
N = 1010
n = 0
dx, dy = [-1, 0, 1, 0], [0, 1, 0, -1]
times = T
g = [[0 for i in range(N)] for j in range(N)]
m = []

def dfs(x, y):
    m[x][y] = 0
    for i in range(4):
        a, b = x, y
        a += dx[i]
        b += dy[i]
        if a >=0 and a < R and b >= 0 and b < C and m[a][b] == 1:
            dfs(a, b)

while(T):
    R, C = map(int, input().split())
    ans = []
    for i in range(R):
        buf = input().strip()
        for j in range(C):
            g[i][j] = int(buf[j])

    n = int(input())
    for patch in range(n):
        mark = input().split()
        if mark[0] == 'Q':
            res = 0
            m = [row[:C] for row in g[:R]]
            for x in range(R):
                for y in range(C):
                    if m[x][y] == 1:
                        dfs(x, y)
                        res += 1
            ans.append(res)
        elif mark[0] == 'M':
            g[int(mark[1])][int(mark[2])] = int(mark[3])
    print("Case #%d:"% (times - T + 1))
    for i in ans:
        print(i)
    T -= 1
```
