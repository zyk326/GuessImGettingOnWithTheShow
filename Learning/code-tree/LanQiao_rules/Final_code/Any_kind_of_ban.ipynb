```python
# 快速幂
def qmi(x, k, mod):
    res = 1 % mod
    while(k):
        if k & 1:
            res = res * x % mod
        x = x * x % mod
        k >>= 1
    return res

# 快速幂
def qmi(x, k, mod):
    res = 1 % mod
    while(k):
        if k & 1:
            res = res * x % mod
        x = x * x % mod
        k >> 1
    return res
```
